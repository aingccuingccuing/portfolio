```python
# distance 추가

import networkx as nx
import matplotlib.pyplot as plt

WG = nx.Graph()

# add relationship and its value
city_graph = city_graph = [('Chennai', 'Delhi', {'distance':700}), ('Mumbai', 'Chennai',{'distance':300}), ('Delhi', 'Kolkata', {'distance':1000}),
               ('Delhi', 'Bangalore',{'distance':20}), ('Chennai', 'Hyderabad',{'distance':100}), ('Chennai', 'Jaipur',{'distance':1200})]

# edges are the lines
WG.add_edges_from(city_graph)

# on the edge, add the value of the relationship (in dict) using the key, "distance"
labels = nx.get_edge_attributes(WG, "distance")

# need layout for the format
pos = nx.spring_layout(WG)

#draw with the position layout
nx.draw(WG,pos,with_labels=True)

# add edges(values of the relationship, relationship is not explicit)
#nx.draw_networkx_edge_labels(WG, pos, edge_labels=labels)

plt.show()
```

# 추가 개념 설명 (node, relationship 등은 제외)

- Degree (노드 차수): 연결선 개수

- Degree Centrality (차수 중심성): 연결선 개수에 따른 '상대적' 중요도

- Betweenness Centrality (매개 중심성): 최단 거리에서 나타나는 노드 집계, 최단 거리 이동 시 무조건 거쳐 가야하는 노드일 수록 필수적이라고 볼수도 있다.

- Connectivity (연결성): graph가 연결 되어있는지 여부 (고립된 부분 존재여부)

- Connected Components (연결된 구성요소): 서로 연결 되어 있는 구획들

- Diameter (지름): 최단 거리 중 최장 거리. 즉, 모든 노드를 연결하기 위해 요구되는 최소 거리라고 할 수 있다.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges = [('A','B'), ('A','C'), ('B','D'),('B','E'),('C','F'),('C','G')]
G.add_edges_from(edges)

degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.3f}")

betweenness_centrality = nx.betweenness_centrality(G)
print("\nBetweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"{node}: {centrality:.3f}")

degree = dict(G.degree())
print("\nDegree:")
for node, deg in degree.items():
    print(f"{node}: {deg}")

is_connected = nx.is_connected(G)
print("\nIs Conncted:", is_connected)

components = list(nx.connected_components(G))
print("\nConnected Components:", components)

diameter = nx.diameter(G)
print("\nDiameter:", diameter)
```
